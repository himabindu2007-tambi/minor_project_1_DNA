```python
with open('/content/hostel_bois[1].txt', 'r', encoding='utf-8') as file:
    lines = file.readlines()

print("Total Lines:", len(lines))
print(lines[:5])
```

```python
messages = []

system_count = 0
media_count = 0
deleted_count = 0

for line in lines:

    line = line.strip()

    if line == "":
        continue

    if " - " not in line:
        continue

    parts = line.split(" - ", 1)

    timestamp = parts[0]
    rest = parts[1]

    if ": " not in rest:
        system_count += 1
        continue

    sender, text = rest.split(": ", 1)

    if text == "<Media omitted>":
        media_count += 1

    if text == "This message was deleted":
        deleted_count += 1

    messages.append({
        "timestamp": timestamp,
        "sender": sender,
        "text": text
    })

print("Messages Parsed:", len(messages))
print("System Messages:", system_count)
print("Media Messages:", media_count)
print("Deleted Messages:", deleted_count)
```

Feature 2 – Group Overview

```python
participants = set()

for msg in messages:
    participants.add(msg["sender"])

print("Participants:", len(participants))
print(participants)
```

```python
message_count = {}

for msg in messages:

    sender = msg["sender"]

    if sender not in message_count:
        message_count[sender] = 0

    message_count[sender] += 1

sorted_people = sorted(
    message_count.items(),
    key=lambda x: x[1],
    reverse=True
)

print("MESSAGES PER PERSON")

for person, count in sorted_people:
    print(person, ":", count)
```

```python
first_date = messages[0]["timestamp"]
last_date = messages[-1]["timestamp"]

print("First Message:", first_date)
print("Last Message:", last_date)
```

Feature 3 – Most Active Day

```python
day_count = {}

for msg in messages:

    date_part = msg["timestamp"].split(",")[0]

    if date_part not in day_count:
        day_count[date_part] = 0

    day_count[date_part] += 1

busiest_day = max(day_count, key=day_count.get)

print("Busiest Day :", busiest_day)
print("Messages :", day_count[busiest_day])
```

```python
hour_count = {}

for msg in messages:

    time_part = msg["timestamp"].split(",")[1].strip()

    hour = time_part.split(":")[0]

    if hour not in hour_count:
        hour_count[hour] = 0

    hour_count[hour] += 1

busiest_hour = max(hour_count, key=hour_count.get)

print("Busiest Hour :", busiest_hour + ":00")
print("Messages :", hour_count[busiest_hour])
```

Feature 4 – NumPy Activity Heatmap

```python
import numpy as np

people = list(participants)

person_index = {}

for i in range(len(people)):
    person_index[people[i]] = i

heatmap = np.zeros((len(people), 24), dtype=int)

for msg in messages:

    sender = msg["sender"]

    time_part = msg["timestamp"].split(",")[1].strip()

    hour = int(time_part.split(":")[0])

    row = person_index[sender]

    heatmap[row][hour] += 1

print(heatmap)
```

```python
print("ACTIVITY HEATMAP")

for person in people:

    row = heatmap[person_index[person]]

    maximum = max(row)

    print(f"{person:<10}", end=" ")

    for value in row:

        if maximum == 0:
            symbol = ". "

        elif value <= maximum * 0.25:
            symbol = ". "

        elif value <= maximum * 0.50:
            symbol = "░ "

        elif value <= maximum * 0.75:
            symbol = "▒ "

        else:
            symbol = "█ "

        print(symbol, end="")

    print()
```

Feature 5 – Top Words

```python
stop_words = [
    "i","is","the","a","an","and",
    "or","to","of","in","on","for"
]

word_count = {}

for msg in messages:

    text = msg["text"]

    if text == "<Media omitted>":
        continue

    if text == "This message was deleted":
        continue

    words = text.lower().split()

    for word in words:

        word = word.strip(".,!?()[]{}\"'")

        if word in stop_words:
            continue

        if word == "":
            continue

        if word not in word_count:
            word_count[word] = 0

        word_count[word] += 1

top_words = sorted(
    word_count.items(),
    key=lambda x: x[1],
    reverse=True
)

print("TOP 10 WORDS")

for word, count in top_words[:10]:

    bar = "█" * (count // 20)

    print(f"{word:<15} {bar} {count}")
```

Feature 6 – Response Speed

```python
from datetime import datetime

response_times = {}

for person in participants:
    response_times[person] = []

for i in range(1, len(messages)):

    current = messages[i]
    previous = messages[i - 1]

    if current["sender"] == previous["sender"]:
        continue

    t1 = datetime.strptime(
        previous["timestamp"],
        "%d/%m/%y, %H:%M"
    )

    t2 = datetime.strptime(
        current["timestamp"],
        "%d/%m/%y, %H:%M"
    )

    gap = (t2 - t1).total_seconds() / 60

    response_times[current["sender"]].append(gap)

avg_response = {}

for person in participants:

    times = response_times[person]

    if len(times) > 0:
        avg_response[person] = sum(times) / len(times)
    else:
        avg_response[person] = 0

fastest = min(avg_response, key=avg_response.get)
slowest = max(avg_response, key=avg_response.get)

print("Fastest Replier :", fastest)
print("Average :", round(avg_response[fastest], 2), "minutes")

print()

print("Slowest Replier :", slowest)
print("Average :", round(avg_response[slowest], 2), "minutes")
```

```python
from datetime import datetime

all_dates = set()

for msg in messages:
    date = msg["timestamp"].split(",")[0]
    all_dates.add(date)

all_dates = sorted(
    all_dates,
    key=lambda d: datetime.strptime(d, "%d/%m/%y")
)

for person in participants:

    active_dates = set()

    for msg in messages:

        if msg["sender"] == person:
            active_dates.add(
                msg["timestamp"].split(",")[0]
            )

    longest = 0
    current = 0

    for day in all_dates:

        if day not in active_dates:
            current += 1
            longest = max(longest, current)

        else:
            current = 0

    print(person, ":", longest, "days")
```

Feature 7 – Personality Archetypes

```python
archetypes = {}

for person in participants:

    person_msgs = []

    for msg in messages:
        if msg["sender"] == person:
            person_msgs.append(msg)

    total = len(person_msgs)

    night = 0

    for msg in person_msgs:

        hour = int(
            msg["timestamp"].split(",")[1]
            .strip()
            .split(":")[0]
        )

        if hour >= 23 or hour <= 4:
            night += 1

    night_percent = (night / total) * 100

    avg_words = sum(
        len(msg["text"].split())
        for msg in person_msgs
    ) / total

    caps = 0

    for msg in person_msgs:

        text = msg["text"]

        if len(text) > 3:

            if text.isupper() or text.count("!") >= 2:
                caps += 1

    caps_percent = (caps / total) * 100

    if night_percent > 60:
        archetypes[person] = "THE NIGHT OWL"

    elif avg_words > 30:
        archetypes[person] = "THE STORYTELLER"

    elif caps_percent > 30:
        archetypes[person] = "THE DRAMA QUEEN"

    else:
        archetypes[person] = "ACTIVE MEMBER"

for person, role in archetypes.items():
    print(person, "→", role)
```
